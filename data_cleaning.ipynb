```python
from pandas import DataFrame
from numpy.random import randn
df = DataFrame(randn(100,4), columns=list('ABCT'))
df.head()
```

```python
# Creating fake values 
df.loc[3, 'A'] = None

from random import choice
df['B'] = df.apply(lambda r: choice(['xx', 'yy', 'zz', 'kk']), axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
df2 = df.copy()

# colA: repace NaN with value
df2['A']= df2.A.fillna(999)

# colB: For each class in B.unique(); associate number
le = LabelEncoder()
df2['B'] = le.fit_transform(df2['B'])

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder.fit(df2[['B']])

from pandas import concat
df_cat = DataFrame(encoder.transform(df2[['B']]).todense(), columns=le.classes_)
print(df_cat.head())
print(df2.head())
df2.merge(df_cat, left_index=True, right_index=True)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
transformed = pca.fit_transform(df2)
print(transformed)
```
