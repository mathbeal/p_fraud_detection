```python
import pandas as pd
from collections import namedtuple
from itertools import product
import matplotlib.pyplot as plt

prefixs = ('train', 'test')
tokens = ('identity', 'transaction')

# locate csv files
get_file_name = lambda prefix, token: f"{prefix}_{token}"
get_file_path = lambda file_name: f"data/{file_name}.csv"

# csv files to dataframes
files = {}
for prefix, token in product(prefixs, tokens):
    filename = get_file_name(prefix, token)
    files[filename] = pd.read_csv(get_file_path(filename))

# keep dataframes files as namedtuple.
files_nt = namedtuple('DF_Files', 'train_transaction test_transaction train_identity test_identity')
df_files = files_nt(**files)
```

```python
train_merged = df_files.train_transaction.merge(df_files.train_identity, how='left', on='TransactionID')
```

```python
# drop columns with too many na values

# replace nan with value 
train_merged.fillna(-999, inplace=True)

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
unwanted = ('isFraud', 'TransactionID', 'TransactionDT')
features = [c for c in train_merged.columns if c not in unwanted]
target = 'isFraud'
cleaned = train_merged.copy()

X = cleaned[features]

# use categorical data on cols not int64, float 64
for col in X.columns:
    if X[col].dtype in ('int64', 'float64'):
        continue
    X.loc[:,col] = X[col].astype('category').cat.codes

n_components = 2
pca = PCA(n_components=n_components)
X1 = pca.fit_transform(X)
print('#'*10)
print(f"{n_components} components explain {100.*pca.explained_variance_ratio_.sum()}% of the information.")
print('#'*10)
pca_df = DataFrame(data=X1)
finalDf = pd.concat([DataFrame(data=X1), cleaned[[target]]], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
cls = LogisticRegression()

test_merged = df_files.test_transaction.merge(df_files.test_identity, how='left', on='TransactionID')
print(finalDf.head())
print(target)
cls.fit(finalDf[[ c for c in finalDf.columns if c in unwanted ]], finalDf[target])

cleaned = test_merged.copy()
X = cleaned[features]

# use categorical data on cols not int64, float 64
for col in X.columns:
    if X[col].dtype in ('int64', 'float64'):
        continue
    X.loc[:,col] = X[col].astype('category').cat.codes

test_merged = df_files.test_transaction.merge(df_files.test_identity, how='left', on='TransactionID')
predictions = cls.predict(test_merged[features])
print(predictions)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

targets = ['isNotFraud', 'isFraud']
colors = ['r', 'g']
for idx, (target, color) in enumerate(zip(targets,colors)):
    indicesToKeep = finalDf['isFraud'] == idx
    ax.scatter(finalDf.loc[indicesToKeep, 0]
               , finalDf.loc[indicesToKeep, 1]
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
